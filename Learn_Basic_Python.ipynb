```python
#Python Lists
#Used to store multiple items in one variable
#Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.
#List is used using square bracket
mylist=["banana","apple","orange"]
print (mylist)
#List items are ordered , changable and allow duplicate values.
#Ordered means list has a definite order and to add new item we have to add at the end of list.
#Changable meaning that we can change, add, and remove items in a list after it has been created.
#Duplicte members means two same items in one list.
#A list can contain different data types:
mylists=["age", 34, True]
thislist = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)
```

```python
#List is a collection which is ordered and changeable. Allows duplicate members.
#Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
#Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.
#Dictionary is a collection which is ordered** and changeable. No duplicate members.
```

```python
#Unpacking a Tuple
fruits=("orange", "apple", "cherry")
(orange,red,green)=fruits
print(orange)
print(red)
print(green)
#Using Asterisk
#If the number of variables is less than the number of values, you can add an * to the variable name and the values will be assigned to the variable as a list:
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
#add an * to the variable name and the values will be assigned to the variable as a list:
```

```python
#Pyhon Sets
#A set is unordered , unchangeable and unindexed.
#Set items are unchangeable, but you can remove items and add new items.
thisset={"mango","banana","apple",2,True}
#It doesnot allow duplicate values.
#Once a set is created, you cannot change its items, but you can remove items and add new items.
print(thisset)
print(len(thisset))
thisset=set(("mango","banana","apple",2,True))
print(thisset)
print(len(thisset))
```
